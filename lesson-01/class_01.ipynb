```python
def check_inplace_add(obj, addition) -> bool:
    id_obj = id(obj)
    obj += addition

    return obj, id_obj == id(obj)
```

```python
s1 = "qwert"
res = check_inplace_add(s1, "123")

print(res)
```

```python
lst1 = [1, 2, 3]
res = check_inplace_add(lst1, [4, 5])

print(res)
```

```python
def is_dup1(nums: list[int]) -> bool:
    return len(set(nums)) != len(nums)

# O(n)
```

```python
def is_dup2(nums: list[int]) -> bool:
    seen = set()
    for n in nums:  # len
        if n in seen:
            return True
        else:
            seen.add(n) # O(1)
    return False

# O(n)
```
