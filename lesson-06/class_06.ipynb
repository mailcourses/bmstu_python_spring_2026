```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def run(self):
        print("run")

    def __getattribute__(self, name):
        print(f"__getattribute__ {name=}")
        return super().__getattribute__(name)

    def __getattr__(self, name):
        print(f"__getattr__ {name=}")
        raise AttributeError(f"no attr={name} in obj")

    def __setattr__(self, name, val):
        print(f"__setattr__ {name=}, {val=}")
        super().__setattr__(name, val)

    def __delattr__(self, name):
        print(f"__delattr__ {name=}")
        super().__delattr__(name)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def run(self):
        print("run")

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, val):
        self._age = val

    def __getattribute__(self, name):
        print(f"__getattribute__ {name=}")
        return super().__getattribute__(name)

    def __getattr__(self, name):
        print(f"__getattr__ {name=}")
        raise AttributeError(f"no attr={name} in obj")

    def __setattr__(self, name, val):
        print(f"__setattr__ {name=}, {val=}")
        super().__setattr__(name, val)

    def __delattr__(self, name):
        print(f"__delattr__ {name=}")
        super().__delattr__(name)
```

```python
class SpecialIterator:
    def __init__(self, limit):
        self.limit = limit
        self.counter = 0
        print("SpecialIterator: init")

    def __iter__(self):
        print("SpecialIterator: iter")
        return self

    def __next__(self):
        print("SpecialIterator: next")
        if self.counter < self.limit:
            self.counter += 1
            return self.counter
        else:
            raise StopIteration
```

```python
for i in SpecialIterator(3):
    print(f"{i=}")
```

```python
class ListIterator:
    
    def __init__(self, lst):
        self.lst = lst
        self.index = 0
        print("ListIterator: init")

    def __iter__(self):
        print("ListIterator: iter")
        return self

    def __next__(self):
        print("ListIterator: next")
        if self.index == len(lst):
            raise StopIteration

        res = lst[self.index]
        self.index += 1
        return res


class CustomList(list):
    def __iter__(self):
        print("List iter")
        return ListIterator(self)
```

```python
def gen_fib(n):
    print("start fib")
    a, b = 0, 1
    while n > 0:
        print("before yield", b)
        yield b
        print("after yield", b)
        a, b = b, a + b
        n -= 1
    return "all fibs are done"
```
