```python
def append_val_valid(n: int, nums: list[int] | None = None) -> list[int]:
    if nums is None:
        nums = []
    nums.append(n)
    return nums
```

```python
nums1 = append_val_valid(10, [3, 7])
print(nums1)
```

```python
nums2 = append_val_valid(10)
print(nums2)
```

```python
nums3 = append_val_valid(10)
print(nums3)
```

```python
nums2 is nums3
```

```python
def append_val_not_valid(n: int, nums: list[int] | None = None) -> list[int]:
    if not nums:
        nums = []
    nums.append(n)
    return nums
```

```python
nums1 = append_val_not_valid(10, [3, 7])
print(nums1)
```

```python
nums2 = append_val_not_valid(10)
print(nums2)
```

```python
orig_nums = []
nums2 = append_val_not_valid(22, orig_nums)
print(nums2)
```

```python
orig_nums is nums2
```

```python
def deco(fn):
    def inner(*args, **kwargs):
        print("before", fn.__name__)
        res = fn(*args, **kwargs)
        print("after", fn.__name__)
        return res
    return inner


@deco
def add_nums(a, b):
    return a + b


def sum_nums(a, b):
    return a + b

sum_nums = deco(sum_nums)
```

```python
add_nums(1, 2)
```

```python
sum_nums(1, 2)
```

```python
def create_deco(koeff):
    def deco(fn):
        def inner(*args, **kwargs):
            print("before", fn.__name__)
            res = fn(*args, **kwargs) + koeff
            print("after", fn.__name__)
            return res
        return inner
    return deco


@create_deco(50)
def add_nums(a, b):
    return a + b


def sum_nums(a, b):
    return a + b

sum_nums = create_deco(900)(sum_nums)
```

```python
add_nums(1, 2)
```

```python
sum_nums(1, 2)
```

```python
import time


def timeit(fn): # fn -> PyObject[id=1, 'fn_sum_nums']
    def inner(*args, **kwargs):
        t1 = time.time()
        res = fn(*args, **kwargs)
        t2 = time.time()
        print("TT", t2 - t1)
        return res
    print(inner, id(inner))  # inner -> PyObject[id=2, 'fn_inner']
    return inner


@timeit
def add_nums(a, b):
    return a + b


def sum_nums(a, b):  # sum_nums -> PyObject[id=1, 'fn_sum_nums']
    return a + b

id_sum_nums = id(sum_nums)
sum_nums = timeit(sum_nums)  # sum_nums -> PyObject[id=2, 'fn_inner']
```
