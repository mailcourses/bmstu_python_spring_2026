```python
class TrivialError(Exception):
    pass


class DangerError(TrivialError):
    pass
```

```python
try:
    raise TrivialError()
except Exception as err:  # isinstance(err, Exception)
    print("Exception", err)
except TrivialError as err:
    print("TrivialError", err)
except DangerError as err:
    print("DangerError", err)
```

```python
def validate_int(val):
    if not isinstance(val, int):
        raise ValueError("val is not int")
```

```python
try:
    raise TrivialError()
except DangerError as err:  # isinstance(TrivialError(), DangerError) == False
    print("DangerError", err)
except TrivialError as err:  # isinstance(TrivialError(), TrivialError) == True
    print("TrivialError", err)
except Exception as err:  # isinstance(TrivialError(), Exception) == True
    print("Exception", err)
finally:
    print("finally")
```

```python
try:
    raise TrivialError()
except DangerError as err:  # isinstance(TrivialError(), DangerError) == False
    print("DangerError", err)
except TrivialError as err:  # isinstance(TrivialError(), TrivialError) == True
    print("TrivialError", err)
    raise
except Exception as err:  # isinstance(TrivialError(), Exception) == True
    print("Exception", err)
finally:
    print("finally")
```

```python
try:
    raise TrivialError()
    print("after_error")
except DangerError as err:  # isinstance(TrivialError(), DangerError) == False
    print("DangerError", err)
except TrivialError as err:  # isinstance(TrivialError(), TrivialError) == True
    print("TrivialError", err)
    raise Exception("WRONG") from err
except Exception as err:  # isinstance(TrivialError(), Exception) == True
    print("Exception", err)
finally:
    print("finally")
```
