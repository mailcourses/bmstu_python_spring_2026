```python
def process_json(
    json_str: str,
    required_keys: list[str] | None = None,
    tokens: list[str] | None = None,
    callback: Callable[[str, str], Any] | None = None,
) -> None:
    ...


# например:
json_str = '{"key1": "Word1 word2", "key2": "word2 word3"}'
required_keys = ["key1", "KEY2"]
tokens = ["WORD1", "word2"]
calls = []

process_json(json_str, required_keys, tokens, lambda key, token: calls.append((key, value)))

# выведет:
# key="key1", token="WORD1"
# key="key1", token="word2"

assert calls == [("key1", "v1"), ("key1", "v2")]
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, val):
        if not isinstance(val, str):
            raise ValueError("name must be string")

        self.__name = val

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, val):
        if not isinstance(val, int) or val < 0:
            raise ValueError("age must be non-negative int")

        self.__age = val


pers = Person("steve", 99)
```

```python
def make_hash_from_password(password):
    return hash(password)


class User:
    def __init__(self, login, password):
        self.__login = login
        self.password_hash = None
        self.password = password

    @property
    def login(self):
        return self.__login

    @property
    def password(self):
        raise AttributeError("Password is write-only")

    @password.setter
    def password(self, plaintext):
        self.password_hash = make_hash_from_password(plaintext)
```

```python
class Student(Person):
    def __init__(self, school, *args, **kwargs):
        self.school = school
        super().__init__(*args, **kwargs)
        self.name = "other_name"
        self.__system_id = f"school_some_id"

    def print(self):
        return f"{self.__system_id=}: {super().print()}"
        


stud = Student("mfti", "student", "stud", 998)
stud.print()
```

```python
class Person:
    def __init__(self, name, login, system_id):
        self.name = name
        self._login = login
        self.__system_id = system_id

    def print(self):
        return f"Person[{self.name=}, {self._login=}, {self.__system_id=}]"

    @property
    def system_id(self):
        return self.__system_id


class Student(Person):
    def __init__(self, school, *args, **kwargs):
        self.school = school
        super().__init__(*args, **kwargs)
        self.name = "other_name"
        self.__system_id = "other_system_id"

    def print(self):
        return f"Student[{self.name=}, {self._login=}, {self.system_id=}]"


stud = Student("mfti", "student", "stud", 998)
stud.print()
```
