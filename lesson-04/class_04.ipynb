```python
from collections import deque


def timeit_last_k(k):
    if not isinstance(k, int) or k < 1:
        raise ValueError("k has to be int > 0")

    def deco(fn):
        last_calls = deque()
        total_s = 0

        @wraps(fn)
        def inner(*args, **kwargs):
            nonlocal total_s

            t1 = time.time()
            try:
                return fn(*args, **kwargs)
            finally:
                t2 = time.time()

                total_s += t2 - t1
                last_calls.append(t2 - t1)

                if len(last_calls) > k:
                    total_s -= last_calls.popleft()
    
                print(f"TT: {t2 - t1}, len={len(last_calls)}, MEAN: {total_s / len(last_calls)}")

            return res
        return inner
    return deco


@timeit_last_k(5)
def add_nums(a, b):
    return a + b


@timeit_last_k(10)
def sum_nums(a, b):
    return a + b
```

```python
class Num:
    def __init__(self, n):
        self.n = n

    def __add__(self, other):
        print("__add__", self.n)
        return self.n + other

    def __radd__(self, other):
        print("__radd__", self.n)
        return self.n + other
```

```python
n1 = Num(5)
n2 = Num(8)
```

```python
n1 + 11
```
